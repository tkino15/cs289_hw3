# 4. Eigenvectors of the Gaussian Covariance Matrix

```python
import numpy as np
import matplotlib.pyplot as plt
```

## (a)

```python
np.random.seed(289)
x1 = 3 + 3 * np.random.randn(100)
x2 = x1 / 2 + 4 + 2 * np.random.randn(100)

X = np.vstack((x1, x2))
mu = X.mean(axis=1)

print(f'mean:\n{mu}')
```

## (b)

```python
sigma = np.cov(X)

print(f'covariance matrix:\n{sigma}')
```

## (c)

```python
eig_val, eig_vec = np.linalg.eig(sigma)

print(f'eigenvectors:\n{eig_vec}')
print(f'eigenvalues:\n{eig_val}')
```

## (d)

```python
def plot_mvn(x1, x2, eig_val=None, eig_vec=None):
    fig, ax = plt.subplots(figsize = (5, 5))
    ax.scatter(x1, x2, s=8)
    if eig_val is not None and eig_vec is not None:
        ax.arrow(mu[0], mu[1], eig_val[0] * eig_vec[0, 0], eig_val[0] * eig_vec[1, 0], width=0.1)
        ax.arrow(mu[0], mu[1], eig_val[1] * eig_vec[0, 1], eig_val[1] * eig_vec[1, 1], width=0.1)
    ax.set_xlim(-15, 15)
    ax.set_ylim(-15, 15)
    ax.set_xlabel('x1')
    ax.set_ylabel('x2')
    ax.grid()
    
    return fig, ax
```

```python
fig, ax = plot_mvn(x1, x2, eig_val, eig_vec)
```

## (e)

```python
X_nor = X - mu.reshape(-1, 1)
X_rot = np.dot(eig_vec.T, X_nor)
```

```python
fig, ax = plot_mvn(X_rot[0], X_rot[1])
```
